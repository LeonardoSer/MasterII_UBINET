```python
from sklearn.linear_model import LinearRegression
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12.0, 9.0)
```

```python
def linear_regression(X, Y, L, epochs):
    m = 0
    c = 0
    n = float(len(X)) # Number of elements in X

    intermediate_errs = [] # store the error for each step

    # Gradient Descent 
    for i in range(epochs): 
        Y_pred = [m*x + c for x in X]  # The current predicted set of values
        
        step_err = error(Y, Y_pred)
        intermediate_errs.append(step_err)
        
        D_m = (-2/n) * sum([x * (y - y_pred) for x,y,y_pred in zip(X, Y, Y_pred)])  # Derivative wrt m
        m = m - L * D_m  # Update m
        
    return m, c, intermediate_errs

def stochastic_linear_regression(X, Y, L, epochs, batch_size):
    m = 0
    c = 0
    n = float(len(X)) # Number of elements in X

    intermediate_errs = [] # store the error for each step

    # Gradient Descent 
    for i in range(epochs): 
        
        indxs = random.sample([i for i in range(len(X))], int(batch_size))
        X = [X[i] for i in indxs]
        Y = [Y[i] for i in indxs]
        
        Y_pred = [m*x + c for x in X]  # The current predicted set of valuse
        
        step_err = error(Y, Y_pred)
        intermediate_errs.append(step_err)
        
        D_m = (-2/n) * sum([x * (y - y_pred) for x,y,y_pred in zip(X, Y, Y_pred)])  # Derivative wrt m
        m = m - L * D_m  # Update m
        
    return m, c, intermediate_errs
```

```python
# computes the error between real and predicted data
def error(Y, Y_pred):
    tot_err = 0
    for y, y1 in zip(Y, Y_pred):
        tot_err += pow(y-y1, 2)
    return tot_err
```

```python
X = [tmp_x[0] for tmp_x in (np.random.rand(150, 1))]
Y = [ 3 * x + random.random() for x in X]
plt.scatter(X,Y)
```

```python
L = 0.01# The learning Rate
epochs = 1000  # The number of iterations to perform gradient descent

### batch ###
m, c, batch_step_errs = linear_regression(X, Y, L, epochs)

Y_batch = [m*x + c for x in X] # predicted values for Y
batch_err = error(Y, Y_batch) # total error on prediction

### mini batch ###
batch_size = 0.2*len(X) # choose the batch size
indxs = random.sample([i for i in range(len(X))], int(batch_size))
X_b = [X[i] for i in indxs] 
Y_b = [Y[i] for i in indxs]

m, c, mini_batch_step_errs = linear_regression(X_b, Y_b, L, epochs)

Y_mini_batch = [m*x + c for x in X] # predicted values for Y
mini_batch_err = error(Y, Y_mini_batch) # total error on prediction

### stochastic ###
m, c, stochastic_step_errs = stochastic_linear_regression(X, Y, L, epochs, batch_size)

Y_stochastic = [m*x + c for x in X] # predicted values for Y
stochastic_err = error(Y, Y_stochastic) # total error on prediction
```

```python
plt.title("Different Linear Regression approaches")
plt.scatter(X, Y, color="black") 
plt.plot([min(X), max(X)], [min(Y_batch), max(Y_batch)], label="batch, err = "+str(batch_err), color="red")  # regression line
plt.plot([min(X), max(X)], [min(Y_mini_batch), max(Y_mini_batch)], label="mini batch, err = "+str(mini_batch_err), color ="blue")  # regression line
plt.plot([min(X), max(X)], [min(Y_stochastic), max(Y_stochastic)], label="stochastic, err = "+str(stochastic_err), color="orange")  # regression line
plt.legend()
plt.show()
```

```python
plt.title("Progression of the error at each step of the gradient descent for each approach")
plt.plot([step for step in range(epochs)], batch_step_errs, color="red", label="batch intermediate errors")
plt.plot([step for step in range(epochs)], mini_batch_step_errs, color="blue", label="mini batch intermediate errors")
plt.plot([step for step in range(epochs)], stochastic_step_errs, color="orange", label="stochastic intemediate errors")
plt.legend()
plt.show()
```
