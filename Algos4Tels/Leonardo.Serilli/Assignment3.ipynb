```python
def draw(G, node_color, edge_color, pos):
    node_lab = {i for i in G.nodes()}
    plt.figure(figsize=(4,4))
    nx.draw(G, pos, with_labels=True, node_color=node_color, edge_color=edge_color, node_size=500)
    plt.show()

class args:
    filename = None
    scale = 40
    figsize = (5,5)
    
def DrawSol(G, f, K, ST):
    for k in K: 
        plt.figure(figsize=args.figsize)
        print("========== path of commodity " + str(k) + " => (sk, tk) = (" + str(ST[k][0]) + "," + str(ST[k][1]) + ") ==========")
        print()
        pos = nx.shell_layout(G)

        nx.draw_networkx_nodes(G, pos=pos, node_shape='o', node_size=600, node_color='lightblue', label=[G.nodes()])
        nx.draw_networkx_labels(G, pos=pos, font_color='w', font_size=12)


        color = ['#%06X' % random.randint(0, 0xFFFFFF) for i in K]
        all_sol_edges = []

        edges_in_sol = []
        for u,v in G.edges():
            if f[u,v,k].x > 0.01:
                edges_in_sol.append((u,v))
            if f[v,u,k].x > 0.01:
                edges_in_sol.append((v,u))



#        for u,v in G.edges():
#            if((u,v) not in all_sol_edges):
#                nx.draw_networkx_edges(G, pos=pos, edgelist=[(u,v)], edge_color='darkgrey')   
        table = []
        for u,v in edges_in_sol:
            nx.draw_networkx_edges(G, pos=pos, edgelist=[(u,v)], edge_color=color[k])
            table.append([(u,v), f[u,v,k].x])
            
        print(tabulate(table, headers=["arc","flow"]))
        print()
        plt.axis('off')
        plt.show()
```

# Q2 
A) The minimum capacity c needed to route all the requests is 5

B) Increasing the capacity the cost doesn't change

```python
G = nx.cycle_graph([i for i in range(1, 11)])
G.add_edges_from([(2,4), (5,10), (6,9)])
G = G.to_directed()
draw(G, "lightblue", "black", nx.shell_layout(G))

ST = [(1,2), (1,6), (1,7), (2,3), (3,10), (2,7), (3,8), (3,9), (7,6), (9,2)] 
K = [i for i in range(len(ST))]
B = [1, 2, 1, 1, 3, 2, 1, 1, 1, 2]

alpha = 100
beta = 5
gamma = 20
capacity = 5


multi = gb.Model()

x = multi.addVars(G.edges(), vtype=gb.GRB.BINARY, name='x')
f = multi.addVars(G.edges(), K, vtype=gb.GRB.CONTINUOUS, name='f')

multi.setObjective(gb.quicksum(alpha*x[i,j] for i,j in G.edges()) \
                   + gb.quicksum(beta*f[i,j,k] for i,j in G.edges() for k in K) \
                   + gb.quicksum(gamma*x[i,j]*x[j,i] for i,j in G.edges()))

# Balance constraint  
for v in G.nodes():
    for k in K:
        if v == ST[k][0]:
            multi.addConstr(f.sum('*',v,k) - f.sum(v,'*',k) == -B[k])
        elif v == ST[k][1]:
            multi.addConstr(f.sum('*',v,k) - f.sum(v,'*',k) == B[k])
        else: 
            multi.addConstr(f.sum('*',v,k) - f.sum(v,'*',k) == 0)
    
all_edges = []
for i,j in G.edges():
    all_edges.append((i,j))
    all_edges.append((j,i))
    
for i,j in G.edges():
    multi.addConstr(gb.quicksum(f[i,j,k] for k in K) <= capacity) # Capacity Constraint
    for k in K:
        multi.addConstr(f[i,j,k] >= 0)


multi.update()
multi.optimize()

print("\n\n the optimal value is ", multi.objval, "\n\n")

G = G.to_undirected()
DrawSol(G, f, K, ST)
```
