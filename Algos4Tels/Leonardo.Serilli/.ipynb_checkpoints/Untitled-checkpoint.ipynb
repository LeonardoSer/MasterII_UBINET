```python
import networkx as nx
import gurobipy as gb
from itertools import combinations, chain
import pygraphviz as pygv
import os
from IPython.display import SVG, display
import matplotlib.pyplot as plt
```

```python
def draw(G, node_color, edge_color, pos):
    node_lab = {i for i in G.nodes()}
    plt.figure(figsize=(4,4))
    nx.draw(G, pos, with_labels=True, node_color=node_color, edge_color=edge_color, node_size=500)
    plt.show()
```

```python
G = nx.cycle_graph([i for i in range(1, 11)])
G.add_edges_from([(2,4), (5,10), (6,9)])
G = G.to_directed()
draw(G, "lightblue", "black", nx.shell_layout(G))
```

```python
def multi_comm_sens_model(ST, K, B, capacity):
    multi = gb.Model()

    x = multi.addVars(G.edges(), K, vtype=gb.GRB.BINARY, name='x')
    q = multi.addVars(G.nodes(), K, vtype=gb.GRB.CONTINUOUS, name='q')
    f = multi.addVars(G.edges(), K, vtype=gb.GRB.CONTINUOUS, name='f')

    multi.setObjective(x.sum())

    # Balance constraint  
    for v in G.nodes():
        for k in K:
            if v == ST[k][0]:
                multi.addConstr(f.sum('*',v,k) - f.sum(v,'*',k) == -1)
            elif v == ST[k][1]:
                multi.addConstr(f.sum('*',v,k) - f.sum(v,'*',k) == 1)
            else: 
                multi.addConstr(f.sum('*',v,k) - f.sum(v,'*',k) == 0)

    for i,j in G.edges():
        multi.addConstr(gb.quicksum(f[i,j,k] for k in K) <= capacity) # Capacity Constraint
        multi.addConstr(gb.quicksum(x[i,j,k] for k in K) <= 1) 

        for k in K:
            multi.addConstr(f[i,j,k] >= 0)
            multi.addConstr(q[j,k] >= q[i,k] + 1 - len(G.nodes()) * (1 - x[i,j,k]))

    for k in K:
        multi.addConstr(gb.quicksum(x[i,j,k] for i,j in G.edges()) >= 1)

    multi.update()
    multi.Params.LogToConsole = 0
    multi.optimize()

    for v in x:
        print(x[v])
    return multi.objval 
```

```python
ST = [(1,2), (1,6), (1,7), (2,3), (3,10), (2,7), (3,8), (3,9), (7,6), (9,2)] 
K = [i for i in range(len(ST))]
B = [1, 2, 1, 1, 3, 2, 1, 1, 1, 2]
```

```python
for capacity in [6]:
    print("The objective value for capacity", capacity, " is ->", multi_comm_sens_model(ST, K, B, capacity))
```
